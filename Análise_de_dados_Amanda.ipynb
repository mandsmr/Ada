#Este relatório apresenta uma análise exploratória de dados e busca compreender e identificar o perfil dos clientes que deixaram determinado Banco. Abaixo, seguem as traduções referentes à cada item da coluna, a fim de esclarecer melhor o entedimento dos dados expostos.


CustomerId: IDENFICAÇÃO DO CLIENTE

Surname: SOBRENOME DO CLIENTE

CreditScore: PONTUAÇÃO DE CRÉDIO (0 SIGNIFICA ALTO RISCO DE INADIMPLÊNCIA E 1000 INDICA CLIENTES COM BAIXO RISCO DE INADIMPLÊNCIA)

Geography: PAÍS EM QUE O SERVIÇO É OFERECIDO

Gender: SEXO DO CLIENTE

Age: IDADE DO CLIENTE

Tenure: INDICATIVO DE ESTABILIDADE NO EMPREGO (0 INDICA POUCA ESTABILIDADE E 10 INDICA MUITA ESTABILIDADE)

Balance: SALDO DA CONTA CORRENTE

NumOfProducts: NÚMERO DE PRODUTOS BANCÁRIOS ADQUIRIDOS

HasCrCard: TEM CARTÃO DE CRÉDITO OU NÃO (1=SIM / 0=NÃO)

IsActiveMember: É UM CLIENTE COM CONTA ATIVA (1=ATIVO)

EstimatedSalary: SALÁRIO ESTIMADO

Exited: USUÁRIO DEIXOU DE SER CLIENTE NO BANCO OU NÃO (1=SIM)












#Inicialmente, faz-se necessário importar os dados com o pacote Pandas.

```python
import pandas as pd
```

#Importando o Dataset:

```python
#dataframe equivale a uma tabela de banco de dados
df = pd.read_csv('/content/churn.csv',delimiter=';')
df
```

#Nesse contexto, é importante fazer uma avaliação geral dos dados, a fim de identificar quais itens podem ser descartados, facilitando, assim, o processo de análise.





```python
arr_delete = ['CustomerId', 'Surname'] #foram eliminadas as colunas 'Idenficação do cliente', 'Sobrenome do cliente'
df.drop(arr_delete, axis = 1, inplace = True) #axis 1 refere-se a coluna
df
```

#No banco há um total de 10000 usuários





```python
df['Exited'].value_counts() #Entre os 10000 usuários, 2037 estão entre os que se ausentaram do banco
```


#É fundamental averiguar e distinguir os dados em qualitativos e quantitativos.

```python
df.info() 
```

```python
#separar as Features qualitativas e quantitativas
#definir um array

arr_quali = ['Geography', 'Gender', 'HasCrCard', 'IsActiveMember', 'Exited']
arr_quanti = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary', 'Exited'] #exited está presente em ambos porque é a variável alvo

df_quali = df[arr_quali]
df_quanti = df[arr_quanti]
```

```python
df_quali
```

```python
df_quanti
```

#É válido também observar os dados (arredondados) e definir, dentre eles, quais fazem, de fato, sentido para o objetivo estabelecido inicialmente.

```python
df_quanti.describe().round(0) 
#25% dos dados totais ganham mais de 51.0000
#50% dos dados totais ganham mais de 100.0000
#75% dos dados totais ganham mais de 149.0000
```

#A seguir algumas análises pontuais serão feitas a fim de traçar melhor o perfil desejado.

```python
df_quanti.loc[df_quanti.Age == 18] #tem duas pessoas com 18 anos que apresentam confiança máxima
```

```python
df_quanti.loc[(df_quanti.Age == 18) & (df_quanti.Tenure >= 5)] #aqui foram observadas pessoas com idade igual a 18 anos e uma estabilidade maior/igual a 5
```

#Ao selecionar os clientes com idade igual a 18 anos e estabilidade baixa (menor que 5), notou-se que essas pessoas não saíram do banco. Logo, não interessa a análise.

```python
df_quanti.loc[(df_quanti.Age == 18) & (df_quanti.Tenure < 5)] 
```

```python
df_quanti.loc[(df_quanti.EstimatedSalary < 1000)] #salário anual menor que 1000
```

```python
df.loc[(df.EstimatedSalary == 440.20) | (df.EstimatedSalary == 11.58)] #análises mais específicas entre clientes com salários distintos
```

```python
df.loc[(df.EstimatedSalary == 90.07) | (df.EstimatedSalary == 91.75)] #análises mais específicas entre clientes com salários muito próximos
```

#Dando prosseguimento, é válido apontar que clientes com salários baixos e com uma quantia alta de dinheiro no banco tendem a deixá-lo.

```python
df_quanti.loc[(df_quanti.EstimatedSalary < 1000) & df.Exited == 1] #pessoas que apresentam o salário baixo e saíram do banco
```

```python
df_quanti.loc[(df_quanti.EstimatedSalary < 1000) & (df_quanti.Exited == 0) & (df_quanti.Balance > 0)]
#pessoas que tem mais chance de sair do banco, pois se assemelham ao perfil daquelas que já saíram
```

#A tabela abaixo mostra os clientes que, de fato, deixaram o banco.

```python
df.loc[df['Exited'] == 1] 
```

#É importante ressaltar a quantidade de pessoas que saíram do banco. Sob essa ótica, nota-se que dentre os 10000 usuários, 2037 se retiraram, o que equivale a 20,37% do total de clientes.

```python
df_1= df.query('Exited == 1')
df_2 = df_1['Exited']
df_2.value_counts()
```

#Abaixo segue o gráfico comparativo entre as pessoas que permaneceram e as que saíram do banco.

```python
import seaborn as sns
sns.countplot(data = df, x = 'Exited');
```

#Nessa lógica, com o intuito de estabelecer o perfil dessas pessoas que se ausentaram do banco, a seguir serão feitas análises entre as variáveis quantitativas e qualitativas em relação às pessoas que saíram. No quesito localização geográfica, por exemplo, percebe-se que a maior parte da população encontra-se na Alemanha, seguido da França e, em menor proporção, a Espanha, conforme ilustrado no gráfico abaixo.





```python
df_1= df.query('Exited == 1')
df_2 = df_1[['Geography', 'Exited']]
df_2.value_counts()
```

```python
sns.countplot('Geography', hue = 'Exited', data = df); #gráfico que representa a localização dos clientes que saíram do banco
```

#Com relação ao sexo dos indivíduos, houve mais evasão de mulheres.

```python
df_1= df.query('Exited == 1')
df_2 = df_1[['Gender', 'Exited']]
df_2.value_counts()
```

```python
sns.barplot(data=df, x = 'Gender', y='Exited', ci=None); #gráfico que representa o sexo dos clientes que saíram do banco
```

#É possível, ainda, estabelecer a relação ente o 'sexo' e a 'localização' dos clientes do banco em relação à saída dos usuários.

```python
import seaborn as sns
sns.barplot(data=df, x='Geography', y='Exited', hue='Gender', ci=None);
```

#Com relação à pontuação de crédito, percebe-se que há clientes com Score alto (850) que deixaram o banco.

```python
df_1= df.query('Exited == 1')
df_2 = df_1[['CreditScore', 'Exited']]
df_2.value_counts()
```

#Além disso, todos os clientes que possuem pontuação de crédito menor que 400 saíram do banco

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,12))
sns.barplot(data=df[(df["CreditScore"] <=850) & (df["CreditScore"] < 400)], x = "CreditScore", y = "Exited");
```

#Já em relação à idade dos clientes, grande parte dos usuários concentram-se na faixa etária entre 40 e 65 anos.

```python
plt.figure(figsize = (20,12))
sns.barplot(data=df[(df["Age"] >=15) & (df["Age"] < 86)], x = 'Age', y='Exited', ci = None); #gráfico que representa a idade dos clientes do banco
```

#Quanto ao grau de estabilidade dos clientes, a maioria dos clientes que saíram do banco possuem pouca estabilidade (Tenure = 1).

```python
df_1= df.query('Exited == 1')
df_2 = df_1[['Tenure', 'Exited']]
df_2.value_counts()
```

```python
sns.countplot('Tenure', hue = 'Exited', data = df); #gráfico que representa o grau de estabilidade dos clientes que saíram do banco
```

#O 'Balance' representa o saldo disponível na conta do cliente. Nesse sentido, nota-se, pela análise dos dados, que a maioria das pessoas que saíram possuem um saldo zerado (Balance = 0).  

```python
df_1= df.query('Exited == 1')
df_2 = df_1[['Balance', 'Exited']]
df_2.value_counts()
```

#Sobre os produtos adquiridos pelos usuários, observa-se que 1409 entre as 2037 das pessoas que saíram adquiriram 1 produto.

```python
df_1= df.query('Exited == 1')
df_2 = df_1[['NumOfProducts', 'Exited']]
df_2.value_counts()
```

```python
sns.countplot('NumOfProducts', hue = 'Exited', data = df); #gráfico que representa o número dos produtos adquiridos pelos clientes que saíram do banco
```

#Observando a estimativa salarial, a evasão dos clientes ocorreu independente do salário, tendo sido constatado que clientes saíram com salários baixos e altos.



```python
df_1= df.query('Exited == 1')
df_2 = df_1[['EstimatedSalary', 'Exited']]
df_2.value_counts()
```

#Acerca do cartão de crédito, nota-se que a maioria das pessoas que se retiraram do banco possuem cartão de credito ['HasCrCard' = 1 (possui cartão)]

```python
df_1= df.query('Exited == 1')
df_2 = df_1[['HasCrCard', 'Exited']]
df_2.value_counts()
```

```python
sns.countplot('HasCrCard', hue = 'Exited', data = df);
```

#Por vim, percebe-se que a maioria dos usuários que se ausentaram não possuem uma conta ativa no banco [IsActiveMember = 0 (não possui conta ativa)].

```python
df_1= df.query('Exited == 1')
df_2 = df_1[['IsActiveMember', 'Exited']]
df_2.value_counts()
```

```python
sns.countplot('IsActiveMember', hue = 'Exited', data = df);
```

- A maior parte dos usuários são da Alemanha (39,96%) 
- A maior parte dos usuários são mulheres (55,91%) 
- Foi registrada evasão de usuários com Score baixo (menor que 400) e Score alto (até 850)
- A maior parte dos usuários encontra-se com idade entre 40 e 65 anos
- Grande parte dos usuários apresenta baixa estabilidade (por exemplo: 11,39% possuem Tenure = 1)
- Grande parte dos usuários apresentam um saldo zerado no banco (24,54% possuem Balance = 0) 
- A maior parte dos usuários adquiriram apenas 1 produto (69,17%) 
- Com relação ao Salário, notou-se evasão de clientes com salários baixos e altos
- A maioria dos usuários possuem cartão de crédito (69,90%) 
- A maioria dos usuários não possuem conta ativa no banco (63,92%)
